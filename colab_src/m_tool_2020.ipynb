```
import math

def get_permutation(n, r):
    """ 順列の総数を返す関数
    異なるn個のものからr個を選んだ並べ方
    """
    return math.factorial(n) // math.factorial(n - r)


def get_combination(n, r):
    """ 組合せの総数を返す関数
    順序を考えないで選ぶことを「組合せ」
    nCr=npr/r!
    """
    return get_permutation(n, r) // math.factorial(r)



def main():
    print("順列: " + str(get_permutation(10, 3)))
    print("組合せ: " + str(get_combination(4, 2)))


if __name__ == "__main__":
    main()
```

```
import random
import itertools

# 食事
def do_meal(price, time_hour):
    happiness = 100 * price * time_hour
    return happiness

# 制約
def remain_hp(used_money, used_time):
    remain_money = 28 - used_money
    remain_time = 24 * 30 - used_time
    return [remain_money, remain_time]

# 快速電車　確率
def bk_get_good_train():
    # print(random.uniform(100, 200))

    element_list = ['特快', '快速', '各停']
    prob_list = [0.1, 0.3, 0.6]
    sample_size = 100

    samples = random.choices(element_list, k = sample_size, weights = prob_list)

    # 結果の確認
    for i in element_list:
        print(i, ':', samples.count(i))


def get_good_train():
    element_list = ['特快', '快速', '各停']
    prob_list = [0.1, 0.3, 0.6]
    sample_size = 1

    samples = random.choices(element_list, k = sample_size, weights = prob_list)
    print(samples)

    if "快速" == samples[0]:
        return 10
    else:
        return 2

def get_perm():
    elem_list = ["食事", "テレビ", "お風呂", "洗濯", "睡眠"]
    comb = itertools.combinations(elem_list, 3)
    return list(comb)

def main():
    print("食事: " + str(do_meal(0.06, 1)))
    print("制約: " + str(remain_hp(0.06, 1)))
    print("電車確率: " + str(get_good_train()))
    print(get_perm())

    list_mo = []
    for str_m in get_perm():
        for moji in str_m:
            list_mo.append(moji)

    print(list(map(lambda str_moji : str_moji + "test", list_mo)))

if __name__ == "__main__":
    main()
```

```
import itertools


act_list = ['お風呂', '睡眠', '食事', 'テレビ', 'ギター', 'ウクレレ']
happiness_list = [30, 20, 50, 15, 40, 30]
money_list = [100, 0, 700, 100, 10, 10]
time_list = [0.5, 4, 0.5, 3, 1, 0.5]

decision_dic = dict(zip(act_list, zip(happiness_list, money_list, time_list)))

print(decision_dic)

cnt = 0
max_fx = -100000
max_combi = ""
max_happiness = 0
max_money = 0
max_time = 0
for sel_num in range(len(act_list)):
    comb_list = list(itertools.combinations(act_list, sel_num + 1))
    for acts in comb_list:
        cnt +=1
        happiness_sum = 0
        money_sum = 0
        time_sum = 0
        fx = 0
        for act in acts:
            happiness_sum += decision_dic[act][0]
            money_sum += decision_dic[act][1]
            time_sum += decision_dic[act][2]
        # fx = 1 * happiness_sum - abs(250-money_sum) - abs(6-time_sum)
        fx = 1.5 * happiness_sum/sum(happiness_list) - money_sum/sum(money_list) \
             - time_sum/sum(time_list)
        if fx > max_fx:
            max_fx = fx
            max_combi = acts
            max_happiness = happiness_sum
            max_money = money_sum
            max_time = time_sum

print(cnt)
print(max_fx)
print(max_combi)
print(max_happiness)
print(max_money)
print(max_time)
```

```
from pyqubo import Binary, solve_qubo

# １日の時間
one_day_time = 5

a = Binary("仕事")           # 8時間 
b = Binary("お風呂")         # 0.5時間
c = Binary("ご飯")           # 1時間
d = Binary("洗濯")           # 0.5時間
e = Binary("掃除")           # 0.5時間
f = Binary("お勤め1(IT系)")   # １時間
g = Binary("お勤め2(ギター)") # 2時間
h = Binary("テレビ")          # 1時間
i = Binary("twitter")         # 1時間
j = Binary("Youtube")         # 1時間
k = Binary("睡眠")            # 8時間
l = Binary("日々の悩み")      # 1時間

H = (one_day_time - a*8 - b*0.5 - c*1 - d*0.5 - e*0.5 - f*1 - g*2 - h*1 - i*1 - j*1 - k*8 -l*1)**2

qubo, offset = H.compile().to_qubo()
solution = solve_qubo(qubo)
print(solution)
```
